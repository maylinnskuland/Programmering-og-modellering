# Numerisk derivasjon

```python
import pylab as pl
```

Newtons kvotient:

```python
h = 0.0000000000001         #delta x
analytisk = 8

def f(x):
    return 2*x**2 + 4*x - 32

def f_derivert(x):
    return (f(x+h)-f(x))/h
```

Varierer dx og lager en tabell med dx og feil

```python
liste_logfeil = []

for i in range(1,17):
    h = 10**-i
    numerisk = f_derivert(1)
    feil = abs(analytisk - numerisk)
    liste_logfeil.append(pl.log10(feil))
    
    print(h, feil)
```

```python
liste_logfeil = []
liste_logh = []

for i in range(1,17):
    h = 10**-i
    liste_logh.append(pl.log10(h))
    numerisk = f_derivert(1)
    feil = abs(analytisk - numerisk)
    liste_logfeil.append(pl.log10(feil))
    

x = liste_logh
y = liste_logfeil

pl.xlabel('10^x')
pl.ylabel('feilen') 

pl.grid('on')

pl.plot(x,y,'c-')
pl.show()
```

Det som skjer er at i starten vil feilen bli stor fordi dx ikke er liten nok og det blir overslagsfeil. Etterhvert som man gjør dx mindre og mindre vil feilen bli mindre, helt til man kommer til 10^-9. Da vil feilen bli større og større igjen. Vi kan derfor se at det er best å sette 


$∆x = 10^{-8}$

### Bruker filen bevelgelse.txt for å plotte fart etter en viss tid

```python
tid = []
fart = []
fil = open('bevegelse.txt', 'r')

for linje in fil:
    data = linje.split(',')
    print(data)
    tid.append(float(data[0]))
    fart.append(float(data[1]))

fil.close

```

```python
x = tid
y = fart

pl.xlabel('Tid')
pl.ylabel('Hastighet') 

pl.grid('on')

pl.plot(x,y,'c-')
pl.show()
```

### Deriverer et bilde:

Starter med å laste inn bildet

```python
bilde = pl.imread("bilde.png")
pl.imshow(bilde)
pl.savefig("orginal.png")
pl.show
```

Deriverer alle punktene vannrett for hverandre

```python
dfdx = abs(bilde[1:-1,2:,:] - bilde[1:-1,1:-1,:])
pl.imshow(10 * dfdx)
pl.show()
pl.savefig("dfdx.png")
```

Deriverer alle punktene loddrett fir hverandre

```python
dfdy = abs(bilde[2:,1:-1,:] - bilde[1:-1,1:-1,:])
pl.imshow(10 * dfdy)
pl.show()
pl.savefig("dfdy.png")
```

Setter til slutt de to bildene sammen

```python
grad = pl.sqrt(dfdx**2 + dfdy**2)
pl.imshow(10 * grad)
pl.savefig("grad.png")
```

Oppdaget av hvis jeg fant summen av de to nærliggende punktene (i stedenfor differansen), så fikk jeg hevet lysstyrekn på bildet

```python
dfdy = abs(bilde[:,:,:] + bilde[:,:,:])
pl.imshow(dfdy)
pl.show()
```
