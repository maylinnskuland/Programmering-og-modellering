# Differensiallikninger - Populasjon av stubbneseaper

```python
import pylab as pl
```

Programmet under sier hvor mange stubbneseaper det er i en populasjon. Jeg starter med 28 aper i populasjonen, og etter tre år har populasjonen vokst med 0.01 aper per måned. Nederste er det en graf som viser en diskret vekst av aper.

```python
p_0 = 28        
k = 0.01
aar = 3
t = aar * 12        #tid i månder
P = p_0

aper = []
tid = []


for i in range(t):
    P = P + k*P
    aper.append(P)
    tid.append(i)
print("Etter", aar, "år er det nå", round(P,0), "stubbneseaper i populasjonen")

x = tid
y = aper

pl.xlabel('Tid i måneder')
pl.ylabel('Populasjon') 

pl.grid('on')

pl.plot(x,y,'b-')
pl.show()
```

```python
x = tid
y = aper

pl.xlabel('Tid i måneder')
pl.ylabel('Populasjon') 

pl.grid('on')

pl.plot(x,y,'r.')
pl.show()
```
